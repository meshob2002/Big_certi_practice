***
**<span style = "color:red">Questions : 데이터에서 IQR을 활용해 Fare컬럼의 이상치를 찾고, 이상치 데이터의 여성 수를 구하시오</span>**


````python
import seaborn as sns
df = sns.load_dataset('titanic')
df
````

````python
df['fare'].describe()
````

````python
# 내답안
Q1 = df['fare'].quantile(.25)
Q3 = df['fare'].quantile(.75)
IQR = Q3 - Q1
low_sigma = Q1 - 1.5*IQR
high_sigma = Q3 + 1.5*IQR
x = (df['fare']< low_sigma )| (df['fare']> high_sigma)
df[x]
````

````python
# 이상치 데이터의 수와 여성의 수
print(f'전체 요금중 이상치의 수 : {len(df[x])}')
print(f"이상치 여성의 수 : {sum(df[x]['who'] == 'woman')}")
````

````python
# 답안 코드
# IQR 구하기
Q1 = df['fare'].quantile(.25)
Q3 = df['fare'].quantile(.75)

# numpy 활용시
# Q1 = np.percentile(df['fare'], 25)
# Q3 = np.percentile(df['fare'], 75)

IQR = Q3 - Q1
Q1 - 1.5 * IQR, Q3 + 1.5 * IQR
# 이상치 데이터 구하기
outdata1 = df[df['fare']<(Q1 - 1.5 * IQR)]
outdata2 = df[df['fare']>(Q3 + 1.5 * IQR)]
print(len(outdata1), len(outdata2))
# 이상치 데이터에서 여성 수 구하기, 출력하기 print()
print(sum(outdata2['who'] == 'woman'))
````

###  주어진 데이터에서 이상치 확인
소수점 올림 np.ceil()  
소수점 내림 np.floor()  
반올림 round()

***
**<span style = "color:red">Questions :
주어진 데이터에서 이상치(소수점 나이)를 찾고 올림, 내림, 버림(절사)했을때 3가지 모두 이상치 'age' 평균을 구한 다음 모두 더하여 출력하시오</span>**

````python
import numpy as np
df = pd.read_csv('basic1.csv')
df['age'].describe()
````

````python
# 내 답안
x = df['age'].apply(str) 
a =[]
for i in x :
    if i[-1] != "0" :
        a.append(i)

a = list(map(float,a))
ceil = list(map(np.ceil,a))
floor = list(map(np.floor,a))
trunc = list(map(np.trunc,a))
````

````python
print(np.mean(ceil))
print(np.mean(floor))
print(np.mean(trunc))
print(f'총합: {np.mean(ceil)+np.mean(floor)+np.mean(trunc)}')
````

````python
# 답안 코드
# 소수점 데이터 찾기
df = df[(df['age']-np.floor(df['age']))!= 0]
df
````

````python
# 이상치를 포함한 데이터 올림, 내림, 버림의 평균값 

# 올림
m_ceil = np.ceil(df['age']).mean()

# 내림
m_floor = np.floor(df['age']).mean()

# 버림
m_trunc = np.trunc(df['age']).mean()

# 평균값 더한 다음 출력
print(m_ceil + m_floor + m_trunc)
````

###  왜도와 첨도




DataFrame.skew() 왜도  
DataFrame.kurt() 첨도  
np.log1p() 로그 스케일링
***
**<span style = "color:red">Questions :
주어진 데이터 중 train.csv에서 'SalePrice'컬럼의 왜도와 첨도를 구한 값과, 'SalePrice'컬럼을 스케일링(log1p)로 변환한 이후 왜도와 첨도를 구해 모두 더한 다음 소수점 2째자리까지 출력하시오</span>**

````python
df = pd.read_csv('train.csv')
df
````

````python
# 내답안
sk = df['SalePrice'].skew()
ku = df['SalePrice'].kurt()
print("첨도는: ",sk) # 첨도의 수치가 높을수록 그래프가 뾰족하고, 낮을수록 완만한 모양
print("왜도는: ",ku) # 왜도가 양수일때 오른쪽 꼬리 (mode-median-mean) 애도가 음수일때 (mean-median-mode)
#Kline(2005)은 왜도의 기준을 절대값 3을 초과하지 않고, 첨도는 절대값 8또는 10을 초과하지 않으면 정규분포로 볼 수 있다고 제시하였습니다.
````

````python
df['SalePrice'] = df['SalePrice'].apply(np.log1p)
log_sk = df['SalePrice'].skew()
log_ku = df['SalePrice'].kurt()
print("스케일링 후 첨도는: ",log_sk) # 첨도의 수치가 높을수록 그래프가 뾰족하고, 낮을수록 완만한 모양
print("스케일링 후 왜도는: ",log_ku) # 왜도가 양수일때 오른쪽 꼬리 (mode-median-mean) 애도가 음수일때 (mean-median-mode)
````

````python
print(f'총합 : {sk+ku+log_sk+log_ku : 0.2f}')
````

이때 발생하는 누적합 결측치는 바로 뒤의 값을 채우고, 누적합의 평균값을 출력한다. (단, 결측치 바로 뒤의 값이 없으면 다음에 나오는 값을 채워넣는다)</span>**

````python
#내 답안
df = pd.read_csv('basic1.csv')
df[df['f2'] == 1]
````

````python
x = df[df['f2'] == 1]['f1'].reset_index(drop=True)
k= [0]*(len(x)+1)
for i,j in enumerate(x) :
    if np.isnan(j) :
        k[i+1] = k[i]
        continue
    else :
        k[i+1] =k[i] + x[i]
````

````python
x['f1'] = k[1:]
````

````python
# 시험환경 세팅 (코드 변경 X)
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

def exam_data_load(df, target, id_name="", null_name=""):
    if id_name == "":
        df = df.reset_index().rename(columns={"index": "id"})
        id_name = 'id'
    else:
        id_name = id_name
    
    if null_name != "":
        df[df == null_name] = np.nan
    
    X_train, X_test = train_test_split(df, test_size=0.2, random_state=2021)
    
    y_train = X_train[[id_name, target]]
    X_train = X_train.drop(columns=[target])

    
    y_test = X_test[[id_name, target]]
    X_test = X_test.drop(columns=[target])
    return X_train, X_test, y_train, y_test 
    
df = pd.read_csv("titanic_train.csv")
X_train, X_test, y_train, y_test = exam_data_load(df, target='Survived', id_name='PassengerId')

X_train.shape, X_test.shape, y_train.shape, y_test.shape
````

````python
def char_cate(df, column_name, dic, default_value):
    a = df.copy()
    cate = dic
    for key, value in cate.items():
        start, end = map(int, key.split('~'))
        a.loc[(a[column_name] >= start) & (a[column_name] < end), column_name] = value
    a.loc[~a[column_name].isin(cate.values()), column_name] = default_value
    return a
X_train =  char_cate(X_train,'capital.gain',{'0~0':0,'1~5000':1,'5000~10000':2,'10000~15000':3,'15000~20000':4},5)
X_test = char_cate(X_test,'capital.gain',{'0~0':0,'1~5000':1,'5000~10000':2,'10000~15000':3,'15000~20000':4},5)
X_train = char_cate(X_train,'capital.loss',{'0~0':0,'1~1000':1,'1000~2000':2,'2000~3000':3,'3000~4000':4},5)
X_test = char_cate(X_test,'capital.loss',{'0~0':0,'1~1000':1,'1000~2000':2,'2000~3000':3,'3000~4000':4},5)
````

````python
#label incoding 진행
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
for i in cat_features: 
    X_train[cat_features] = X_train[cat_features].apply(label.fit_transform)
    X_test[cat_features] = X_test[cat_features].apply(label.fit_transform)
````

````python
# IQR을 이용한 이상치 제거 (너무 많이 없어짐, 제거후 데이터와 제거전 데이터셋 비교 필요)
def del_IQR(df,y,column) :
    Q_1 = df[column].quantile(0.25)
    Q_3 = df[column].quantile(0.75)
    IQR = Q_3 - Q_1
    upper = Q_3 + 1.5*IQR
    lower = Q_1 - 1.5*IQR
    upper_idx = df[df[column]>upper].index
    lower_idx = df[df[column]<lower].index
    df = df.drop(upper_idx)
    df = df.drop(lower_idx)
    y = y.drop(upper_idx)
    y = y.drop(lower_idx)
    return df, y
for i in numeric_columns :
    X_train_IQR, y_train_IQR = del_IQR(X_train, y_train, i)
````
